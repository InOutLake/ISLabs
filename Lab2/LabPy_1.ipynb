```python
from experta import *

class KnowledgeBase(KnowledgeEngine):
    def __init__(self):
        super().__init__()
        # Confidence measures
        self.CMGeek = 0 
        self.CMNerd = 0
        self.CMJerk = 0
        # Unconfidence measures
        self.UCMGeek = 0
        self.UCMNerd = 0
        self.UCMJerk = 0

    # DRY purpose
    # def adjust_param_shortlife(self, coef, param, typename):
    #     import pdb
    #     pdb.set_trace()
    #     calc_var = getattr(self, param)[typename]
    #     calc_var = calc_var + coef * (1 - calc_var)
    #     getattr(self, param)[typename] = calc_var
        
    # def base_type_detected(self, coefCM, coefUCM, typename):
    #     caller_name = inspect.currentframe().f_back.f_back.f_code.co_name
    #     logger.info(f'Calling function: {caller_name}')
    #     self.adjust_param_shortlife(coefCM, 'CM', typename)
    #     for t in self.UCM:
    #         if t != typename:
    #             self.adjust_param_shortlife(coefUCM, 'UCM', t)
    
    # def light_type_detected(self, typename):
    #     self.base_type_detected(0.2, 0.15, typename)

    # def advanced_type_detected(self, typename):
    #     self.base_type_detected(0.4, 0.3, typename)

    # def ultra_type_detected(self, typename):
    #     self.base_type_detected(1, 1, typename)

    # def or_fact_connector(facts_answrs_strs: list):
    #     res = Fact(facts_answrs_strs[0])
    #     for ans in facts_answrs_strs[1:]:
    #         res |= Fact(answ=ans)
    #     return res

    # @Rule(or_fact_connector(['1-1', '4-3', '4-2']))
    # def geek_light_type_detected(self):
    #      self.light_type_detected('Geek')
    # @Rule(or_fact_connector(['3-1']))
    # def geek_advanced_type_detected(self):
    #         self.advanced_type_detected('Geek')
    # @Rule(or_fact_connector(['2-1']))
    # def geek_ultra_type_detected(self):
    #         self.ultra_type_detected('Geek')

    # @Rule(or_fact_connector(['1-2', '2-2', '2-3', '3-3', '4-1', '4-3']))
    # def nerd_light_type_detected(self):
    #         self.light_type_detected('Nerd')
    # @Rule(or_fact_connector(['3-2']))
    # def nerd_advanced_type_detected(self):
    #         self.advanced_type_detected('Nerd')

    # @Rule(or_fact_connector(['4-4']))
    # def jerk_light_type_detected(self):
    #         self.light_type_detected('Jerk')
    # @Rule(or_fact_connector(['3-4']))
    # def jerk_advanced_type_detected(self):
    #         self.advanced_type_detected('Jerk')
    # @Rule(or_fact_connector(['1-4', '2-4']))
    # def jerk_ultra_type_detected(self):
    #         self.ultra_type_detected('Jerk')

    @Rule(
        Fact(answ="1-1") | Fact(answ="4-3") | Fact(answ="4-2"))
    def geek_light_type_detected(self):
        self.CMGeek = self.CMGeek + 0.2 * (1 - self.UCMGeek)
        self.UCMNerd = self.UCMNerd + 0.15 * (1 - self.UCMNerd)
        self.UCMJerk = self.UCMJerk + 0.15 * (1 - self.UCMJerk)
    @Rule(Fact(answ="3-1"))
    def geek_advanced_type_detected(self):
        self.CMGeek = self.CMGeek + 0.4 * (1 - self.CMGeek)
        self.UCMNerd = self.UCMNerd + 0.3 * (1 - self.UCMNerd)
        self.UCMJerk = self.UCMJerk + 0.3 * (1 - self.UCMJerk)

    @Rule(Fact(answ="2-1"))
    def geek_ultra_type_detected(self):
        self.CMGeek = 1.0
        self.UCMNerd = 1.0
        self.UCMJerk = 1.0

    @Rule(
        Fact(answ="1-2") | Fact(answ="2-2") | Fact(answ="2-3") |
        Fact(answ="3-3") | Fact(answ="4-1"))
    def nerd_light_type_detected(self):
        self.CMNerd = self.CMNerd + 0.2 * (1 - self.CMNerd)
        self.UCMGeek = self.UCMGeek + 0.15 * (1 - self.UCMGeek)
        self.UCMJerk = self.UCMJerk + 0.15 * (1 - self.UCMJerk)

    @Rule(Fact(answ="3-2"))
    def nerd_advanced_type_detected(self):
        self.CMNerd = self.CMNerd + 0.4 * (1 - self.CMNerd)
        self.UCMGeek = self.UCMGeek + 0.3 * (1 - self.UCMGeek)
        self.UCMJerk = self.UCMJerk + 0.3 * (1 - self.UCMJerk)

    @Rule(Fact(answ="4-4"))
    def jerk_light_type_detected(self):
        self.CMJerk = self.CMJerk + 0.2 * (1 - self.CMJerk)
        self.UCMGeek = self.UCMGeek + 0.15 * (1 - self.UCMGeek)
        self.UCMNerd = self.UCMNerd + 0.15 * (1 - self.UCMNerd)

    @Rule(Fact(answ="3-4"))
    def jerk_advanced_type_detected(self):
        self.CMJerk = self.CMJerk + 0.4 * (1 - self.CMJerk)
        self.UCMGeek = self.UCMGeek + 0.3 * (1 - self.UCMGeek)
        self.UCMNerd = self.UCMNerd + 0.3 * (1 - self.UCMNerd)

    @Rule(Fact(answ="1-4") | Fact(answ="2-4"))
    def jerk_ultra_type_detected(self):
        self.CMJerk = 1.0
        self.UCMGeek = 1.0
        self.UCMNerd = 1.0


    def getCertanities(self):
        return {
            'Geek': self.CMGeek - self.UCMGeek,
            'Nerd': self.CMNerd - self.UCMNerd,
            'Jerk': self.CMJerk - self.UCMJerk
        }
                    
    def checkType(self):
        sorted_certanities = sorted(self.getCertanities(), key=lambda item: item[1])
        return sorted_certanities[0][:2] + sorted_certanities[1][2:]
```

```python
KBase = KnowledgeBase()
KBase.reset()

for ThisKey in Questions:
    print(Questions[ThisKey])
    x = str(input(Answers[ThisKey]))
    ans = str(ThisKey) + "-" + x
    KBase.declare(Fact(answ=ans))
    KBase.run()

print(f"Your type is: {KBase.checkType()}\nCertanities: {KBase.getCertanities()}")
```
